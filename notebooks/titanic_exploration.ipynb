# **Titanic Data Exploration**

This notebook performs initial exploratory data analysis on the cleaned Titanic dataset  
to understand key patterns and relationships before training a model.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv("../data/processed/titanic_clean.csv")

# Display first few rows
df.head()
```

```python
# Distribution of Passenger Class
plt.figure(figsize=(6,4))
df['Pclass'].value_counts().sort_index().plot(kind='bar')
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Survival rate by Sex
surv_by_sex = df.groupby('Sex')['Survived'].mean()
surv_by_sex.index = ['Male', 'Female']

plt.figure(figsize=(6,4))
surv_by_sex.plot(kind='bar')
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Proportion')
plt.show()
```

```python
# Age distribution for survivors vs. non-survivors
plt.figure(figsize=(8,4))
df[df['Survived']==1]['Age'].plot(kind='hist', alpha=0.6, bins=20, label='Survived')
df[df['Survived']==0]['Age'].plot(kind='hist', alpha=0.6, bins=20, label='Did Not Survive')
plt.title('Age Distribution by Survival Outcome')
plt.xlabel('Age')
plt.legend()
plt.show()
```

## **Key Findings**

- **Class Distribution**: Majority of passengers were in 3rd class.  
- **Sex & Survival**: Females had a much higher survival rate than males.  
- **Age & Survival**: Younger passengers show slightly higher survival probability,  
  though age overlaps across both groups.
